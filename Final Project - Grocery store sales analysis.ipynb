1. Який портрет нашого найлояльнішого покупця (вік, стать)?

```python
most_common_combination = df.groupby(['Gender', 'Age']).size().idxmax()
most_common_gender, most_common_age = most_common_combination

print("Портрет нашого найлояльнішого покупця має наступний вигляд: Стать = {}, Вік = {}".format(most_common_gender, most_common_age))
```

Бачимо, що 20 річні жінки є найактивнішими покупцями наших магазинів. Але давайте розглянемо детальніше ТОП-10
портретів наших найбільш і найменш лояльних покупців за допомогою графіків

```python
import matplotlib.pyplot as plt

grouped_data = df.groupby(['Gender', 'Age']).size().reset_index(name='Count')

sorted_data = grouped_data.sort_values(by='Count', ascending=False)

top_10_most_active = sorted_data.head(10)
top_10_least_active = sorted_data.tail(10)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.bar(range(len(top_10_most_active)), top_10_most_active['Count'], color='skyblue')
plt.xlabel('Портрет покупця (Gender, Age)', fontsize=12)
plt.ylabel('Кількість покупок', fontsize=12)
plt.title('Топ 10 найлояльніших портретів покупця', fontsize=14)
plt.xticks(range(len(top_10_most_active)), ['{}, {}'.format(gender, age) for gender, age in zip(top_10_most_active['Gender'], top_10_most_active['Age'])], rotation=45)

plt.subplot(1, 2, 2)
plt.bar(range(len(top_10_least_active)), top_10_least_active['Count'], color='lightcoral')
plt.xlabel('Портрет покупця (Gender, Age)', fontsize=12)
plt.ylabel('Кількість покупок', fontsize=12)
plt.title('Топ 10 найменш лояльних портретів покупця', fontsize=14)
plt.xticks(range(len(top_10_least_active)), ['{}, {}'.format(gender, age) for gender, age in zip(top_10_least_active['Gender'], top_10_least_active['Age'])], rotation=45)

plt.tight_layout()
plt.show()
```

Отже, за допомогою візуалізації можемо зрозуміти, що все ж портрет нашого найлояльнішого покупця - це жінки переважно від 16 до 20 років, але якщо виокремити декілька сегментів, то сюди ж можемо віднести жінок від 25 до 47 років загалом. Чоловіки віком від 61 до 74 років наші найменш лояльні покупці. Ці висновки дозволяють нам зрозуміти як будувати свої рекламні кампанії та які групи покупців варто досліджувати чи опитувати при запуску або покращенні товарів.

2. Які локації магазинів є найбільш і найменш популярними за кількістю покупок?

```python
grouped_data = df.groupby(['City', 'Country']).size().reset_index(name='Count')
most_common_place = grouped_data.loc[grouped_data['Count'].idxmax()]
less_common_place = grouped_data.loc[grouped_data['Count'].idxmin()]

print("Найчастіше місце покупок: Місто = {}, Країна = {}, Кількість покупок = {}".format(most_common_place['City'], most_common_place['Country'], most_common_place['Count']))
print("Найрідше місце покупок: Місто = {}, Країна = {}, Кількість покупок = {}".format(less_common_place['City'], less_common_place['Country'], less_common_place['Count']))
```

Тепер візуалізуємо загальну кількість покупок за містом і країною у вигляді пайчарту

```python
labels = grouped_data['City'] + ', ' + grouped_data['Country']
sizes = grouped_data['Count']
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Розподіл покупок за містом і країною')
plt.show()
```

Далеко не факт, що кількість покупок відповідає показникам чистого прибутку за локаціями. Тому варто перевірити наскільки ці локації є прибутковими.

 9. Які бренди купляють найчастіше за різними локаціями?

```python
import matplotlib.pyplot as plt

grouped_data = df.groupby(['City', 'Brand ']).size().reset_index(name='Count')

top_10_brands = grouped_data.groupby('Brand ')['Count'].sum().nlargest(10).index
grouped_data_top_10 = grouped_data[grouped_data['Brand '].isin(top_10_brands)]

plt.figure(figsize=(12, 6))
for brand in top_10_brands:
    brand_data = grouped_data_top_10[grouped_data_top_10['Brand '] == brand]
    plt.scatter(brand_data['City'], brand_data['Count'], label=brand, alpha=0.7)

plt.title('Кількість покупок за ТОП-10 брендами у різних містах')
plt.xlabel('Місто')
plt.ylabel('Кількість покупок')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

За даним графіком бачимо, що ТОП брендами за різними локаціями однозначно є Bonjour та Biogreen - і компанії буде досить цікаво запровадити співпрацю з цими популярними брендами майже на всіх локаціях, а також, ймовірно, просувати більшу кількість товарів вказаних брендів

10. У які дні здійснювались найкращі і найгірші продажі за останній рік?

```python
import matplotlib.pyplot as plt

df_2019 = df[df['Date'].dt.year == 2019]

daily_profit_2019 = df_2019.groupby('Date')['Profit Margin'].sum()

best_day_2019 = daily_profit_2019.idxmax()
worst_day_2019 = daily_profit_2019.idxmin()

plt.figure(figsize=(12, 6))
plt.plot(daily_profit_2019.index, daily_profit_2019.values, marker='o', color='b')
plt.title('Прибуток по дням у 2019 році')
plt.xlabel('Дата')
plt.ylabel('Прибуток')
plt.xticks(rotation=45)
plt.axvline(x=best_day_2019, color='g', linestyle='--', label=f'Найкращий день: {best_day_2019}')
plt.axvline(x=worst_day_2019, color='r', linestyle='--', label=f'Найгірший день: {worst_day_2019}')
plt.legend()
plt.grid(True)
plt.show()
```

Очевидно, що Різдво було і залишається найкращою датою для продажів, тож компанія може активно використовувати цю інформацію для планування ресурсів, акцій, тощо. Найгірші періоди - періоди відпусток або несвяткові звичайні дні. Це теж може бути важливим при врахуванні закупок товарів і плануванні ресурсів.
